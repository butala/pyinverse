```python
import matplotlib.pylab as plt

from pyinverse.grid import RegularGrid
from pyinverse.phantom3 import Phantom3
```

```python
N = 129
grid = RegularGrid.linspace((-1, 1, N), (-1, 1, N))
```

```python
p3 = Phantom3()
```

```python
Y_n90_0 = p3.proj(-90, 0, grid, deg=True)
```

```python
fig, ax = plt.subplots()
im = grid.plot(ax, Y_n90_0)
plt.colorbar(im)
```
